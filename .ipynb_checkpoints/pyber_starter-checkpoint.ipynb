```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns



# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv("city_data.csv")
ride_data_df = pd.read_csv("ride_data.csv")

#drop one duplicate record with same city name, same ride_ids but different # of drivers
city_data_df = city_data_df.drop_duplicates('city', keep = 'first')

#merge
combined_data_df = city_data_df.merge(ride_df, on = 'city')
```

## Bubble Plot of Ride Sharing Data

```python
#DFs for the types of cities
urban = combined_data_df.loc[(combined_data_df["type"] == "Urban")]
suburban = combined_data_df.loc[(combined_data_df["type"] == "Suburban")]
rural = combined_data_df.loc[(combined_data_df["type"] == "Rural")]

# Urban City: Average Fare, Total Rides and Total Drivers
avg_fare_urban = urban.groupby(['city'])['fare'].mean()
total_rides_urban = urban.groupby(['city']).count()['ride_id']
total_drivers_urban = urban.groupby(['city'])['driver_count'].value_counts()

# Suburban City: Average Fare, Total Rides and Total Drivers 
avg_fare_suburban = suburban.groupby(['city'])['fare'].mean()
total_rides_suburban = suburban.groupby(['city']).count()['ride_id']
total_drivers_suburban = suburban.groupby(['city'])['driver_count'].value_counts()

# Rural City: Average Fare, Total Rides and Total Drivers 
avg_fare_rural = rural.groupby(['city'])['fare'].mean()
total_rides_rural = rural.groupby(['city']).count()['ride_id']
total_drivers_rural = rural.groupby(['city'])['driver_count'].value_counts()

# Construction of Scatter Plots
plt.scatter(total_rides_urban, avg_fare_urban,s=total_drivers_urban*15,
            marker ='o', facecolors ="lightcoral", edgecolors='black',alpha = 0.75, label="Urban")

plt.scatter(total_rides_suburban, avg_fare_suburban,s=total_drivers_suburban*15,
            marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.75, label="Suburban")

plt.scatter(total_rides_rural, avg_fare_rural,s=total_drivers_rural*15,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.75, label="Rural")

# Create a legend and Incorporate a text label regarding circle size
Legend = plt.legend(title = 'City Types', frameon = True, edgecolor = 'black')
Legend.legendHandles[0]._sizes = [65]
Legend.legendHandles[1]._sizes = [65]
Legend.legendHandles[2]._sizes = [65]


plt.xlim(0,45)
plt.ylim(18,45)
plt.title("Pyber Ride Sharing Data",{'fontsize':15})
plt.xlabel("Total Number of Rides (Per City)",{'fontsize':11})
plt.ylabel("Average Fare ($)",{'fontsize':11})
plt.gcf().text(0.95, .25, 'NOTE: Circle size correlates with driver count per city.', fontsize=9)
plt.show()


# Save Figure
plt.savefig("ridesharing.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
Fares = combined_data_df.groupby(['type'])['fare'].sum()
# Construction of Pie Chart
plt.title("Percentage(%) of Total Fares By City Type")
labels = ["Rural","Suburban","Urban" ]
colors = ["gold","lightskyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.pie(Fares, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=165)
plt.axis("equal")

# Save & Show Pie Chart
plt.savefig("PieChart_Fares.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Type Percents
Rides = combined_data_df.groupby(['type'])['ride_id'].count()
# Construction of Pie Chart
plt.title("Percentage(%) of Total Rides By City Type")
labels = ["Rural","Suburban","Urban" ]
colors = ["gold","lightskyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.pie(Rides, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=165)
plt.axis("equal")

# Save & Show Pie Chart
plt.savefig("PieChart_Rides.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Type Percents
Drivers = combined_data_df.groupby(['type'])['driver_count'].sum()
# Construction of Pie Chart
plt.title("Percentage (%) of Total Drivers By City Types")
labels = ["Rural","Suburban","Urban" ]
colors = ["gold","lightskyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.pie(Drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=165)
plt.axis("equal")

# Save & Show Pie Chart
plt.savefig("PieChart_Rides.png")
plt.show()
```
